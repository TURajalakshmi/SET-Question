```python
# 1
import numpy as np

delays = np.array([15, 30, 45, 20, 25, 100, 5, 60, 35, 50,120, 80, 10, 75, 90, 200, 55, 40, 70, 150])

percentiles = np.percentile(delays,[10, 25, 50, 75, 90])
print("Percentiles (10th, 25th, 50th, 75th, 90th):", percentiles)

Q1 = percentiles[1]
Q3 = percentiles[3]

IQR = Q3-Q1
print("IQR:", IQR)

Lower_bound = Q1 - 1.5 * IQR
Upper_bound = Q3 + 1.5 * IQR

Outlier = (delays < Lower_bound) | (delays > Upper_bound)
print ("Outlier:", delays[Outlier])

import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=delays)

plt.title("Box Plot for Flight Delays")
plt.xlabel("Flight Delays (minutes)")
plt.show()
```

```python
# 2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

salaries = np.array([30000, 32000, 35000, 37000, 40000, 42000, 45000, 47000, 50000, 55000,
            60000, 62000, 65000, 67000, 70000, 72000, 75000, 80000, 85000, 90000])

mean = np.mean(salaries)
print("Mean Salary:", mean)

median = np.median(salaries)
print("Median Salary:", median)


#Since no salary value is repeated, there is no true mode in this dataset. The reason SciPy returns 30000 is because of how stats.mode() works. 
#All values tie for the highest frequency (1). stats.mode() breaks the tie by returning the smallest value among the tied values, which is 30000.

mode = stats.mode(salaries).mode
print("Mode Salary:", mode)
print("No Mode (all values occur once)")

skewness = stats.skew(salaries)

if skewness > 0:
    print("The data is Right Skewed (Positive Skew)")
elif skewness < 0:
    print("The data is Left Skewed (Negative Skew)")
else:
    print("The data is Symmetric")

sns.histplot(salaries, kde=True)
plt.title("Histogram of Employee Salaries")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
# 3
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = pd.DataFrame ({
    'Student': ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J'],
    'Mathematics': [85, 78, 92, 88, 70, 95, 60, 80, 90, 76],
    'Science': [80, 85, 88, 70, 75, 92, 55, 82, 89, 78],
    'English': [78, 74, 85, 80, 68, 90, 50, 77, 83, 72]
})

# Descriptive Statistics
print(dataset.describe())

# Histogram(graph)
plt.hist(dataset["Mathematics"], bins = 10)
plt.title("Student mark distribution")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()

# Correlation Analysis
corr_mat = dataset[["Mathematics", "Science", "English"]].corr()
print(corr_mat)

# HeatMap(graph)
marks = dataset.set_index('Student')

sns.heatmap(marks[['Mathematics','Science','English']],
            annot=True,
            cmap='YlGnBu')
```

```python
# 4
from scipy import stats

medication_group = [110, 115, 108, 102, 107, 99, 111, 104, 109, 101]
placebo_group = [120, 125, 130, 122, 128, 119, 124, 127, 123, 126]

t_test, p_value = stats.ttest_ind(medication_group, placebo_group)
print("T-Statistic:", t_test)
print("P-Value:", p_value)

alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis: The medication is effective.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
# 5 
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

ratings = np.array([8, 9, 7, 5, 6, 10, 9, 4, 7, 8, 
           6, 9, 10, 5, 8, 7, 6, 9, 10, 7])

mean = np.mean(ratings)
median = np.median(ratings)
mode = stats.mode(ratings).mode
std_dev = np.std(ratings)

print("Mean Rating:", mean)
print("Median Rating:", median)
print("Mode Rating:", mode)
print("Standard Deviation:", std_dev)

sns.histplot(ratings, bins=6, kde=True)
plt.title("Distribution of Customer Statisfaction Rating")
plt.xlabel("Customer Statisfaction Rating")
plt.ylabel("Frequency")
plt.show()
```
